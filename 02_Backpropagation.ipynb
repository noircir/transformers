```python
# The image above (the blue individual partial derivatives): dw = 2*x *(y_predicted - y)


import torch

x = torch.tensor(1.0)
y = torch.tensor(2.0)

w = torch.tensor(1.0, requires_grad=True)

# forward pass and compute the loss
# Pytorch remembers how the tensors are assembled here:

y_hat = w * x         # one node
loss = (y-y_hat)**2   # 2nd and 3rd nodes

loss  # loss = 1 as manually calculated
```

```python
# backward pass

loss.backward()  # computed backwards from the node 3 -> 2 -> 1 , the chain rule derivatives
w.grad  # dloss/dw. need to minimize loss with respect to weights
```
